syntax : set of rules defines the combinations of symbols.
That are considered to be correctly structured programs in a language.
In simpler terms, syntax is about the correct arrangement of words & symbols in a code.

semantics : Interpretations of symbols, characters, & commands in a language. 
It is about what code is supposed to do when it runs.(behavior)
