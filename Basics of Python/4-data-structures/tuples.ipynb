Tuples are ordered collections of items that are Immutable.They are similar to List but their Immutable nature makes the difference.

```python
##Creating Empty Tuples
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
#creation with function
lst = list() # Creating Empty List
print(type(lst))

tple = tuple()
print(type(tple)) # Creating Empty Tuple
```

```python
numbers = tuple([1,2,3,4,5,6,7,8,9])
print(numbers)
print(type(numbers))
```

```python
mixed_tuples = (1, "Hello World", 3.14, True)
print(mixed_tuples)
```

```python
## Accessing tuple elements
numbers
```

```python
print(numbers[0])
print(numbers[-1])
```

```python
numbers[0:4]
```

```python
numbers[0:4]
```

```python
## Tuple operations

conacat_tuples = numbers + mixed_tuples
print(conacat_tuples)
```

```python
mixed_tuples * 3
```

```python
numbers *3
```

```python

# mutable list

list = [1,2,3,4,5,6]
print(list)
list[3] = "maniraj"
print(list)
```

```python
#Immuatable nature of Tuples

print(numbers)

numbers[1] = "maniraj"

print(numbers)
```

```python
# Tuple Method

print(numbers)
print(numbers.count(1))
print(numbers.index(9))
```

```python
#Packing & unpacking Tuples

#PACKING TUPLE
packed_tuple = 1,2.5,"Maniraj"

print(packed_tuple)


#unpacking Tuple
a,b,c = packed_tuple

print(a,b,c)
```

```python
#unpacking with *

numbers = (1,2,3,4,5,6)
first,*middle,last = numbers

print(first)
print(middle)
print(last)
```

```python
# Nested tuples & list

# Nested List

list = [ [1,2,3,4],[6,7,8,9],[1,"Hello",3.14,'c'] ]
list[0][0:3]
```

```python
lst = [ [1,2,3,4],[6,7,8,9],(1,"Hello",3.14,True) ]
lst[0][0:3]
```

```python
nested_tuple = ( (1,2,3), ("a","b","c"), (True, False))

## access the elements inside tuplr
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
#Iteration in tuple

for item in nested_tuple:
    for sub_item in item:
        print(sub_item,end=" ")
    print()
```

Conclusion

Tuples are versatile and useful in many real-world scenarios where an immutable & ordered collection of items is required.
They are commonly used in data structures, function arguments & return values, and as dictionary keys. Understanding how to leverage tuples 
effectively can improve the efficiency & reliability of python code.
