Set

sets are buit-in data structure in python used to store collections of unique items.
They are un-ordered : elements don't follow a specific order, They don't allow duplicate elements.
Sets are useful for membership test, eliminating duplicate entries.
Also useful for maths operation like union, intersection, difference & symmetric difference. 

```python
## creating a set
myset = {1,2,3,4 , "maniraj"}
print(myset)
print(type(myset))
```
