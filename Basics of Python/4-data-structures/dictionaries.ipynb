```python
# Accessing Dictionary elements
print(employee['name'])
print(employee['age'])
print(employee['level'])
#print(employee['last_name']) This will cause error is key is not in Dictionary.

# Accessing using get method
print(employee.get('name'))
print(employee.get('age'))
print(employee.get('level'))
print(employee.get('last_name'))
print(employee.get('last_name','Not Available'))
```

```python
#Modifying the dictionary Elements
#Dictionary are mutable(add,update,delete)
print(employee)
employee['age'] = 27
print(employee)
employee['address'] = 'India'
print(employee)
```

```python
print(employee)
del employee['level'] # deletes the Key & value pair
print(employee)
```

```python
## Dictionary methods
keys = employee.keys()
print(keys)
values = employee.values()
print(values)

#accessing key value pairs using iteration
for x in employee:
    print(f"key={x} value={employee.get(x)}")

#accessing key value pair as tuple
print(employee.items())

```

```python
#shallow copy
employee_copy = employee #The reference of employee is given to employee_copy
print(employee)
print(employee_copy)
```

```python
#Problem in shallow copy
employee['name'] = 'maniraj'
print(employee)
print(employee_copy)

#Problem :  update in variable dictionary is causing update in its shallow copy
#fix for the problem
employee_copy = employee.copy()

employee['name'] = 'lingala maniraj'
print(employee)
print(employee_copy)
```

```python
#Iteration over dictionaries

#accessing using keys
print()
print("------keys-------")
for key in employee:
    print(key)

#accessing using values
print()
print("------values-------")
for value in employee.values():
    print(value)

#accessing using pairs
print()
print("------pairs-------")
for key,value in employee.items():
    print(f'{key}:{value}')

print()
print("------function-------")
#accessing key value pairs Ineffective way
for x in employee:
    print(f"key={x} value={employee.get(x)}")
```

```python
# Nested Dictionaries
employees = {
    "employee1" : {'name' : 'maniraj' , 'age' : 26},
    "employee2" : {'name' : 'venumadhav' , 'age' : 25},
    "employee3" : {'name' : 'saitharun' , 'age' : 26},
}
print(employees['employee1']['name'])
print(employees['employee2']['name'])
```

```python
#Iteration over nested Dictionary
for employee_id,employee_info in employees.items():#tuple de-packing is being used
    print(f'{employee_id}:{employee_info}')
```

```python
#Iterating over nested dictionaries
for employee_id,employee_info in employees.items():
    print(f"{employee_id}:")
    for employee in employee_info.items():
        print(employee,end=" ")
    print()
```

```python
# Dictionary Comprehension

#square = {x : (x+1)**2 for x in range(5)}
#print(square)

#Practical example
#use a dictionary to count the frequency of elements in list
numbers = [1,2,3,4,1,2,3,4]

frequency ={}

for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:frequency[number] = 1
print(frequency)
```
